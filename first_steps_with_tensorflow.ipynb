```python
# creating a tensorflow dataset
a = [1.2, 3.4, 7.5, 4.1, 5.0, 1.0]
ds = tf.data.Dataset.from_tensor_slices(a)
for item in ds:
    print(item)
```

```python
# batching the tensorflow dataset
ds_batch = ds.batch(2)
for i, item in enumerate(ds_batch, 1):
    print(f'batch {i}: {item.numpy()}')
```

```python
# combining two tensors into joint dataset
tf.random.set_seed(2021)
tx = tf.random.uniform((4, 3))
ty = tf.range(4)
ds_x = tf.data.Dataset.from_tensor_slices(tx)
ds_y = tf.data.Dataset.from_tensor_slices(ty)

ds_joint = tf.data.Dataset.zip((ds_x, ds_y))
for data in ds_joint:
    print(f'x: {data[0].numpy()}, y: {data[1].numpy()}')
```

```python
# another way of combining datasets
ds_joint = tf.data.Dataset.from_tensor_slices((tx, ty))
for data in ds_joint:
    print(f'x: {data[0].numpy()}, y: {data[1].numpy()}')
```

```python
# apply transformations on the dataset
ds_trans = ds_joint.map(lambda x, y: (x*2 - 1.0, y))
for data in ds_trans:
    print(f'x: {data[0].numpy()}, y: {data[1].numpy()}')
```

```python
tf.random.set_seed(1)
ds = ds_joint.shuffle(buffer_size=len(tx))
for data in ds:
    print(f'x: {data[0].numpy()}, y: {data[1].numpy()}')
```

```python
# create batches
ds = ds_joint.batch(batch_size=3, drop_remainder=False)
bx, by = next(iter(ds))
print(f'Batch-X:\n{bx.numpy()}')
print(f'Batch-y:\n{by.numpy()}')
```

```python
# create copies of each batch; 
# suitable for iterating the data over certain number of epochs
ds = ds_joint.batch(batch_size=3).repeat(count=2)
for i, (bx, by) in enumerate(ds):
    print(i, bx.shape, by.numpy())
```

```python
# Let's try to understand the operations: (batch, shuffle, repeat)
# Order 1: shuffle -> batch -> repeat, recommended way of doing it!
tf.random.set_seed(1)
ds = ds_joint.shuffle(buffer_size=len(tx)).batch(2).repeat(3)
for i, (bx, by) in enumerate(ds):
    print(i, bx.shape, by.numpy())
```

```python
# Order 2: batch -> shuffle -> repeat: does not work as we expect it to
tf.random.set_seed(1)
ds = ds_joint.batch(2).shuffle(buffer_size=len(tx)).repeat(3)
for i, (bx, by) in enumerate(ds):
    print(i, bx.shape, by.numpy())
```

```python
# Order 3: batch -> repeat -> shuffle: same, does not work as expected
tf.random.set_seed(1)
ds = ds_joint.batch(2).repeat(3).shuffle(buffer_size=len(tx))
for i, (bx, by) in enumerate(ds):
    print(i, bx.shape, by.numpy())
```

```python
# important imports for downloading and processing images
import shutil
from pathlib import Path
from io import BytesIO
import requests

import matplotlib.pyplot as plt
from PIL import Image
%matplotlib inline
```

```python
# shutil.rmtree('images')
```

```python
# create a directory and download all the images from GitHub
path = Path().cwd() / 'images'
path.mkdir(parents=True, exist_ok=True)
image_path = 'https://raw.githubusercontent.com/rasbt/python-machine-learning-book-3rd-edition/master/ch13/cat_dog_images/'
fnames = [name + idx for idx in ['-01.jpg', '-02.jpg', '-03.jpg'] for name in ['cat', 'dog']]
for i in range(len(fnames)):
    r = requests.get(image_path+fnames[i], stream=True)
    if r.status_code == 200:
        r.raw.decode_content = True
        bytes_img = BytesIO(r.content)
        img = Image.open(bytes_img)
        img.save(path / fnames[i])
```

```python
# load all the image path in a list
file_list = sorted([str(p) for p in path.glob('*.jpg')])
file_list
```

```python
# display the images
fig = plt.figure(figsize=(12, 6))
for i, file in enumerate(file_list):
    img = Image.open(file).convert('RGB')
    print(f'Image shape: {img.size}')
    ax = fig.add_subplot(2, 3, i+1)
    ax.set_xticks([])
    ax.set_yticks([])
    ax.imshow(img)
    ax.set_title(Path(file).name, size=16)
plt.tight_layout()
plt.show()
```

```python
# create integer encoded labels from image names
labels = [0 if Path(f).name.split('-')[0] == 'cat' else 1 for f in file_list]
labels
```

```python
# create tf dataset
images_labels_ds = tf.data.Dataset.from_tensor_slices((file_list, labels))
for item in images_labels_ds:
    print(f'{item[0].numpy()}, {item[1].numpy()}')
```

```python
# helper function to load the images and preprocess them
def load_and_preprocess(f_path, label):
    image = tf.io.read_file(f_path)
    image = tf.image.decode_jpeg(image, channels=3)
    image = tf.image.resize(image, (img_height, img_width))
    image /= 255.0
    return image, label
```

```python
# display the preprocessed images
img_width, img_height = 120, 80
images_labels_ds = images_labels_ds.map(load_and_preprocess)

fig = plt.figure(figsize=(12, 6))
for i, example in enumerate(images_labels_ds):
    ax = fig.add_subplot(2, 3, i+1)
    ax.set_xticks([])
    ax.set_yticks([])
    ax.imshow(example[0])
    ax.set_title(f'{example[1].numpy()}', size=16)
plt.tight_layout()
plt.show()
```

```python
# creating dataset of form (image, label)
ds_train = ds_train.map(lambda item: (item['image'], 
                                      tf.cast(item['attributes']['Male'], tf.int32)))
ds_train = ds_train.batch(16)
images, labels = next(iter(ds_train))
print(images.shape, labels)
```
